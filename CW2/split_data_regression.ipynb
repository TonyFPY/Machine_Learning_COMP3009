```python
import pandas as pd
import numpy as np
import datetime
```

```python
# load raw data
df = pd.read_csv('PM2.5.csv')

# clean data with missing values          
'''drop the rows directly -> mess up the order
    first 24 rows have pm2.5 value that is NaN -> discard
    else: forward filling'''
df = df[24:].fillna(method='ffill')
df=df.reset_index()
```

```python
def checkSeason(x):
    if (3<=x<6):
        return 1
    elif (6<=x<9):
        return 2
    elif (9<=x<12):
        return 3
    elif (x==12 or 1<=x<3):
        return 4

df['season']=df['month'].map(lambda x: checkSeason(x))
```

```python
def checkCBWD(x):
    if (x=='SE'):
        return 1
    elif (x=='cv'):
        return 2
    elif (x=='NW'):
        return 3
    elif (x=='NE'):
        return 4

df['cbwd']=df['cbwd'].map(lambda x: checkCBWD(x))
    
```

```python
def checkTime(x):
    if (0<=x<12):
        return 1
    elif (12<=x<24):
        return 2

df['time']=df['hour'].map(lambda x: checkTime(x))
```

```python
df.drop(columns=['year', 'month', 'day', 'hour', 'No','index'], inplace=True)

df
```

```python
import scipy.io as scio

feature = 'regression_feature.mat'
scio.savemat(feature, {'DEWP':np.array(df['DEWP'],dtype=float),'TEMP':np.array(df['TEMP'],dtype=float),'PRES':np.array(df['PRES'],dtype=float),
                       'cbwd':np.array(df['cbwd'],dtype=float),'Iws':np.array(df['Iws'],dtype=float),'Is':np.array(df['Is'],dtype=float),
                       'Ir':np.array(df['Ir'],dtype=float),'season':np.array(df['season'],dtype=float),'time':np.array(df['time'],dtype=float)})

label = 'regression_label.mat'
scio.savemat(label, {'pm2Dot5':np.array(df['pm2.5'],dtype=float)})
```
